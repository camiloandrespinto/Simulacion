```python
%matplotlib inline 

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

N = 1000
def histo(x):
    
    plt.subplot(2,1,1)
    count, bins, ignored = plt.hist(x, 25, normed=True)
    plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
    plt.show()
    
    plt.subplot(2,1,2)
    x1 = np.arange(1/float(N),1+1/float(N),1/float(N))
    y1 = np.cumsum(np.sort(u)/np.max(np.cumsum(u)))
    x2 = np.sort(x)
    y2 = np.cumsum(np.sort(x)/np.max(np.cumsum(x)))
    Te=np.absolute(y2-y1)
    plt.plot(x1,y1,label='Uniforme')
    plt.step(x2,y2,label='Random')
    x3 = [x1[Te.argmax()],x1[Te.argmax()]]
    y3 = [np.min([y1[Te.argmax()],y2[Te.argmax()]]),np.max([y1[Te.argmax()],y2[Te.argmax()]])]
    plt.step(x3,y3,label='D')

    plt.legend()
    plt.xlabel('X')
    plt.ylabel('Probabilidad Acumulada P(X)')
    plt.ylim(0, 1.05)

    plt.show()


#Generador Numpy
u = [1.0] * N    
x = np.random.uniform(0,1,N)
histo(x)
```
