DISTRIBUCION UNIFORME

```python
def funprobun(x):
    fpu=[]
    for i in range(x):
        fpu.append(1/float(x))
    return fpu

fpu=funprobun(x)
   
def funacumun(x):
    facu=[]
    facu.append(fpu[0])
    for i in range(x-1):
        facu.append(fpu[i+1]+facu[i])
    return facu

plt.plot(funacumun(x))
plt.show()
```

BINOMIAL

```python
%matplotlib inline 

import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
import scipy.integrate
n=10
b=[]
p=[]
bu=[]
acum=[]
#generador Aleatorio
def numalea():
    xo=7
    for i in range(n):
        mod=(223*xo+23)%829
        ui=float(mod)/829
        xo=mod
        p.append(ui)
    #print(p)
    return p
plt.subplot(3,1,1)

def binomial():
    p=numalea()
    for i in range(n):
        pr=(1-p[2])**(n-i)
        c=(math.factorial(n)/(math.factorial(i)*math.factorial(n-i)))*(p[2]**i)*pr
        b.append(c)
        #print(c)
    return b
plt.plot(binomial(),color="r",marker="o",linestyle="--")
plt.show()
plt.subplot(3,1,2)
def acumubino():
    g=0
    for i in range(n):
        g=b[i]+g
        acum.append(g)
    #print("Acumulada",acum)
    return acum
plt.plot(acumubino())
plt.show()

plt.subplot(3,1,3)
def inverbino():
    p=numalea()
    for i in range(n):
        for j in range(n):
            if p[i]<acum[j]:
                bu.append(j)
                break
    print("Valores Inversa Binomial",bu)
    return bu
plt.plot(inverbino())
plt.show()
```

GEOMETRICA

```python
ll=[]
acumg=[]
def geometrica():
    p=numalea()
    for i in range(n):
        l=((1-p[2])**(i-1))*p[2]
        ll.append(l)
    #print(ll)
    return ll

plt.subplot(2,1,1)
def acumugeome():
    h=0
    for i in range(n):
        h=ll[i]+h
        acumg.append(h)
    #print("Acumulada",acum)
    return acumg

plt.plot(geometrica(),label="Geometrica",color="r",marker="o",linestyle="--")
plt.plot(acumugeome(), label="Acumulada")
plt.legend(loc="upper left")

plt.subplot(2,1,2)
bj=[]
def invergeo():
    p=numalea()
    for i in range(n):
        for j in range(n):
            if p[i]<acumg[j]:
                bj.append(j)
                break
    print("Valores Inversa Binomial",bj)
    return bj
plt.plot(invergeo(),marker="x")
plt.grid()
plt.show()
```

DISTRIBUCION CAUCHY

```python
x=10
def genale(N):
    Ux=[]
    x=0
    x0=7
    while x<N:
        x0=((5*x0)+3)%607
        x=x+1
        Ux.append(float(x0)/607)
    return Ux

def funcuancau(x,xo,g):
    p=[]
    p=genale(x)
    pi=[]
    for i in range(x):
        pi.append(xo+g*math.tan(math.pi*(p[i]-(1/2))))
    return pi

fcuc=funcuancau(x,0,1)
for i in range(len(fcuc)):
    print "{0:.2f}".format(fcuc[i])
```
